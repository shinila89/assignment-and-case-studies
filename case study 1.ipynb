# 3.Plot a bar graph of male vs female buyers participated in the sales


import pandas as pd
data=pd.read_csv(r'C:\Users\USER\.jupyter\cars_data.csv')
data['Buyer Gender'].value_counts()

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1.4)
data['Buyer Gender'].value_counts().plot(kind='bar',figsize=(7, 6),rot=0,color=['red','blue'])
plt.xlabel("Gender",labelpad=14)
plt.ylabel("count of people",labelpad=14)
plt.title('bar graph of male vs female buyers participated in the sales')
```

# 4.Find the top 5 cars based on their sales price

```python
top_sales_price=data.sort_values(by=['Sale Price']).tail()
top_cars_df=[top_sales_price["Make"]+" "+top_sales_price["Model"],top_sales_price['Sale Price']]
headers=["Car Name","Sale Price"]
top_cars=pd.concat(top_cars_df,axis=1,keys=headers)
```

```python
top_cars
```

# 5.Find the least 5 cars based on their resell price

```python
least_resell_price=data.sort_values(by=['Resell Price']).head()
least_cars_df=[least_resell_price["Make"]+" "+least_resell_price["Model"],least_resell_price['Resell Price']]
headers=["Car Name","Resell Price"]
least_cars=pd.concat(least_cars_df,axis=1,keys=headers)
```
